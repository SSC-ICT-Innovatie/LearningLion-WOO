```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

RESULTS_FILE = "document_similarity_minbzk_no_requests_FastBM25.csv"

woo_data = pd.read_csv(f"./evaluation_ministries_full/results/{RESULTS_FILE}")
woo_data_list = woo_data["retrieved_page_ids"].dropna().tolist()
woo_data_list_splitted = [i.split(", ") for i in woo_data_list]
```

```python
# Get frequency of first item
first_elements = [sublist[0] for sublist in woo_data_list_splitted if sublist]

# Count the frequency of each first element
first_elements_frequency = Counter(first_elements)

# Get the order of frequency (most common first)
first_elements_order = first_elements_frequency.most_common()

print("Frequency:", first_elements_order[:10])
```

```python
labels, values = zip(*first_elements_order[:10])

# Plotting
plt.figure(figsize=(14, 8))
plt.bar(labels, values, color="skyblue")
plt.xlabel("Page id")
plt.ylabel("Frequency")
plt.title("Bar Chart of Page id frequency on the top 1 retrieved pages")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
woo_data_flattened = [x for xs in woo_data_list_splitted for x in xs]
elements_frequency = Counter(woo_data_flattened)

# Get the order of frequency (most common first)
elements_order = elements_frequency.most_common()

print("Frequency:", elements_order[:10])
```

```python
labels, values = zip(*elements_order[:10])

# Plotting
plt.figure(figsize=(14, 8))
plt.bar(labels, values, color="skyblue")
plt.xlabel("Page id")
plt.ylabel("Frequency")
plt.title("Bar Chart of Page id frequency on the top 1 retrieved pages")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
