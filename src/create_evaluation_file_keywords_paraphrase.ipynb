```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama3")
```

```python
import json

prompt_keywords = "In het Nederlands, haal de kernwoorden uit de volgende text. De kernwoorden mogen niet gaan over wob/woo, en moet puur gaan over het verzoek dat in de text staat. De kernwoorden mogen ook niet over de ministerie gaan. Geef me alleen de kernwoorden en helemaal niks anders, gesplitst door komma's die handig zijn om later weer dit document terug te vinden. Ik heb geen andere context. Kondig niet aan dat dit de kern woorden zijn. Geef me maximaal 5 kernwoorden."
prompt_paraphrase = """
Haal de kern uit dit stuk. De text mag niet gaan over wob/woo, en moet puur gaan over het verzoek dat in de text staat. Geef geen context waarom je dit doet, ik wil alleen de kern hebben. Gebruik geen opsomming, geef alleen text. Begin je antwoord met "U verzoekt om..."
"""

def interact_llm(i, data):
    print(f"Now doing {i}")
    result = llm.invoke(f"{prompt_keywords}\n{data[:1000]}")
    # result = llm.invoke(f"{prompt_paraphrase}\n{data[:1000]}")
    print(result)
    return result

ministry = 'minbzk'

with open(f'final_evaluation/{ministry}.json', 'r') as file:
    data = json.load(file)

new_data = {interact_llm(i, key): value for i, (key, value) in enumerate(data.items())}
```

```python
with open (f"final_evaluation/keywords_{ministry}.json", 'w') as outfile:
    json.dump(new_data, outfile)
```
